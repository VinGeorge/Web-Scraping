```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
news_list = []
for year in range(2015,2020):
    news_list.append('https://nplus1.ru/news/' + str(year))
```

```python
URL = 'https://nplus1.ru'
page = requests.get(URL)
```

```python
soup = BeautifulSoup(page.text, 'lxml')
```

```python
news = []
for link in soup.find_all('a'):
    tag = link.get('href')
    if '/news' in tag:
        full_url = 'https://nplus1.ru' + tag
        news.append(full_url)
```

```python
URL0 = news[0]
page0 = requests.get(URL0)
soup0 = BeautifulSoup(page0.text, 'lxml')
```

```python
def get_text(html):
    text = " ".join([t.text for t in html.find_all('p', {'class': None})])
    text = text.replace('\n', ' ')
    text = text.replace('\xa0', ' ')

    return text.split('Нашли опечатку?')[0]
    
```

```python
topic = [rubrick.text for rubrick in soup0.find_all('p')[0].find_all('a')]
diff = soup0.find_all('span', {'class' : 'difficult-value'})[0].text
author = soup0.find_all('meta', {'name': 'author'})[0].attrs['content']
date = soup0.find_all('meta', {'itemprop': 'datePublished'})[0].attrs['content']
title = soup0.find_all('meta', {'property' : 'og:title'})[0].attrs['content']
text = get_text(soup0)
```

```python
def get_news(URL):
    
    page = requests.get(URL)
    soup = BeautifulSoup(page.text, 'lxml')
    
    topic = [rubrick.text for rubrick in soup.find_all('p')[0].find_all('a')]
    diff = soup.find_all('span', {'class' : 'difficult-value'})[0].text
    author = soup.find_all('meta', {'name': 'author'})[0].attrs['content']
    date = soup.find_all('meta', {'itemprop': 'datePublished'})[0].attrs['content']
    title = soup.find_all('meta', {'property' : 'og:title'})[0].attrs['content']
    link = title = soup.find_all('meta', {'property' : 'og:url'})[0].attrs['content']
    text = get_text(soup)
    
    return link, topic, diff, author, date, title, text
```

```python
get_news(URL0)
```
